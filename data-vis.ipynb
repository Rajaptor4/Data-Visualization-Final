```python
import pandas as pd
import altair as alt

file_path = '../input/car-sales-report/Car Sales.xlsx - car_data.csv'
car_sales = pd.read_csv(file_path)

car_sales['Date'] = pd.to_datetime(car_sales['Date'])

# Visualization 1: Top Car Brands by Sales Volume
top_companies = car_sales['Company'].value_counts().head(10).reset_index()
top_companies.columns = ['Company', 'Sales']

#Bar Chart
top_brands_chart = alt.Chart(top_companies).mark_bar().encode(
    x=alt.X('Sales', title='Total Sales'),
    y=alt.Y('Company', sort='-x', title='Car Brand'),
    color='Company'
).properties(
    title='Top 10 Car Brands by Sales Volume'
)

# Visualization 2: Seasonal Sales Trends

car_sales['Month'] = car_sales['Date'].dt.to_period('M')
monthly_sales = car_sales.groupby('Month').size().reset_index(name='Sales')

monthly_sales['Month'] = monthly_sales['Month'].dt.to_timestamp()

seasonal_trends_chart = alt.Chart(monthly_sales).mark_line().encode(
    x=alt.X('Month:T', title='Month'),
    y=alt.Y('Sales:Q', title='Number of Sales'),
    tooltip=['Month:T', 'Sales:Q']
).properties(
    title='Seasonal Sales Trends'
)

top_brands_chart.display()
seasonal_trends_chart.display()
```
